```python
#sigmoid vs normal; for plevel=0     
import matplotlib.pyplot as ax

# Generate data for Normal case
p_init = 0
plevel = 0
pot1 = []
p_initList1 = []
while p_init <= 0.80:
    k = (1 - plevel) * p_init + plevel
    pot1.append(k)
    plevel = k
    p_initList1.append(p_init)
    p_init += 0.05

# Generate data for Sigmoid case
p_init = 0.5
plevel = 0
pot2 = []
p_initList2 = []
while p_init <= 1:
    k = (1 - plevel) * p_init + plevel
    pot2.append(k)
    plevel = k
    p_initList2.append(p_init)
    p_init += 0.05


# Plot and scatter for Normal case
color = "blue"
label = "Normal"
ax.plot(p_initList1, pot1, color=color, linewidth=2)
ax.scatter(p_initList1, pot1, color=color, s=30, label=label, alpha=0.6)

# Plot and scatter for Sigmoid case
color = "red"
label = "Sigmoid"
ax.plot(p_initList2, pot2, color=color, linewidth=2)
ax.scatter(p_initList2, pot2, color=color, s=30, label=label, alpha=0.6)

# Set labels for the x and y axis with increased font size
ax.xlabel("P_init", fontsize=20, fontweight='bold')
ax.ylabel("P_level", fontsize=20, fontweight='bold')

# Set x and y limits
ax.xlim(left=0)
ax.ylim(bottom=0)


# Add grid with a MATLAB-like appearance
ax.grid(True, linestyle='--', alpha=0.7)

# Add legend with custom settings
ax.legend(loc='best', fontsize=14)

# Save the figure with high resolution and ensure all elements are captured
ax.savefig('pmc-plots/p_init_Vs_p_level.eps', format='eps', dpi=300, bbox_inches='tight')

# Show the plot
ax.show()
```
